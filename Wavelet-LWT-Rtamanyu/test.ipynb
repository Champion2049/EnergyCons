```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mine = pd.read_csv('haar_lwt_1D_L3_coeffs_speed_dataset.csv')
mine.dropna(inplace=True)
matl = pd.read_csv('approximation_coefficients_matlab_l3.csv')
plt.plot(mine['Final_Approximation'], label='Our LWT Coefficients')
plt.plot(matl['ca'], label='Matlab Coefficients')
plt.plot(mine['Final_Approximation']*3, label='Our LWT Coefficients * 3 (level)')
plt.legend()
plt.title('Comparison of Our LWT Coefficients with Matlab Level 3')
plt.xlabel('Coefficient Index')
plt.ylabel('Coefficient Value')
# plt.show()
plt.savefig('haar_lwt_1D_L3_coeffs_comparison.png')
plt.show()

mine = pd.read_csv('haar_lwt_1D_L2_coeffs_speed_dataset.csv')
mine.dropna(inplace=True)
matl = pd.read_csv('approximation_coefficients_matlab_l2.csv')
plt.plot(mine['Final_Approximation'], label='Our LWT Coefficients')
plt.plot(matl['ca'], label='Matlab Coefficients')
plt.plot(mine['Final_Approximation']*2, label='Our LWT Coefficients * 2 (level)')
plt.legend()
plt.title('Comparison of Our LWT Coefficients with Matlab Level 2')
plt.xlabel('Coefficient Index')
plt.ylabel('Coefficient Value')
# plt.show()
plt.savefig('haar_lwt_1D_L2_coeffs_comparison.png')
plt.show()

mine = pd.read_csv('haar_lwt_1D_L4_coeffs_speed_dataset.csv')
mine.dropna(inplace=True)
matl = pd.read_csv('approximation_coefficients_matlab_l4.csv')
plt.plot(mine['Final_Approximation'], label='Our LWT Coefficients')
plt.plot(matl['ca'], label='Matlab Coefficients')
plt.plot(mine['Final_Approximation']*4, label='Our LWT Coefficients * 4 (level)')
plt.legend()
plt.title('Comparison of Our LWT Coefficients with Matlab Level 4')
plt.xlabel('Coefficient Index')
plt.ylabel('Coefficient Value')

plt.savefig('haar_lwt_1D_L4_coeffs_comparison.png')
plt.show()
```
